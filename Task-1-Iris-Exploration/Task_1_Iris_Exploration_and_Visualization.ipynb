## 1. Introduction & Problem Statement

The objective of this task is to **explore, summarize, and visualize** the famous Iris dataset. The Iris dataset is one of the most well-known datasets in data science, originally introduced by the British statistician and biologist **Ronald Fisher** in 1936.

The dataset contains **150 samples** from three species of Iris flowers:
- *Iris Setosa*
- *Iris Versicolor*
- *Iris Virginica*

For each sample, four features were measured:
1. **Sepal Length** (cm)
2. **Sepal Width** (cm)
3. **Petal Length** (cm)
4. **Petal Width** (cm)

### Goals:
- Load and inspect the dataset structure
- Perform basic data summarization
- Create visualizations: **Scatter Plots**, **Histograms**, and **Box Plots**
- Draw meaningful conclusions from the analysis

## 6. Exploratory Data Analysis (EDA) — Visualizations

### 6.1 Scatter Plots — Analyzing Relationships Between Variables

```python
# Scatter Plot 1: Sepal Length vs Sepal Width
plt.figure(figsize=(10, 7))
colors = {'setosa': '#FF6B6B', 'versicolor': '#4ECDC4', 'virginica': '#45B7D1'}

for species in df['species'].unique():
    subset = df[df['species'] == species]
    plt.scatter(subset['sepal_length'], subset['sepal_width'], 
                c=colors[species], label=species, s=80, alpha=0.7, edgecolors='black', linewidth=0.5)

plt.xlabel('Sepal Length (cm)', fontsize=14)
plt.ylabel('Sepal Width (cm)', fontsize=14)
plt.title('Scatter Plot: Sepal Length vs Sepal Width', fontsize=16, fontweight='bold')
plt.legend(title='Species', fontsize=12, title_fontsize=13)
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot 2: Petal Length vs Petal Width
plt.figure(figsize=(10, 7))

for species in df['species'].unique():
    subset = df[df['species'] == species]
    plt.scatter(subset['petal_length'], subset['petal_width'], 
                c=colors[species], label=species, s=80, alpha=0.7, edgecolors='black', linewidth=0.5)

plt.xlabel('Petal Length (cm)', fontsize=14)
plt.ylabel('Petal Width (cm)', fontsize=14)
plt.title('Scatter Plot: Petal Length vs Petal Width', fontsize=16, fontweight='bold')
plt.legend(title='Species', fontsize=12, title_fontsize=13)
plt.tight_layout()
plt.show()
```

**Observations from Scatter Plots:**
- **Setosa** is clearly separable from the other two species based on both sepal and petal measurements.
- **Petal Length vs Petal Width** provides better species separation than Sepal Length vs Sepal Width.
- **Versicolor** and **Virginica** have some overlap, especially in sepal measurements.

```python
# Pair Plot — Relationships between ALL variable pairs
print("Pair Plot of All Features:")
palette = {'setosa': '#FF6B6B', 'versicolor': '#4ECDC4', 'virginica': '#45B7D1'}
pair_plot = sns.pairplot(df, hue='species', palette=palette, 
                         plot_kws={'alpha': 0.7, 's': 60, 'edgecolor': 'black', 'linewidth': 0.3},
                         diag_kind='kde', height=2.5)
pair_plot.figure.suptitle('Pair Plot: All Feature Relationships by Species', y=1.02, fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

### 6.2 Histograms — Examining Data Distribution

```python
# Histograms for all numerical features
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Distribution of Iris Features (Histograms)', fontsize=18, fontweight='bold', y=1.01)

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
titles = ['Sepal Length (cm)', 'Sepal Width (cm)', 'Petal Length (cm)', 'Petal Width (cm)']
color_list = ['#FF6B6B', '#4ECDC4', '#45B7D1']

for idx, (feature, title) in enumerate(zip(features, titles)):
    ax = axes[idx // 2, idx % 2]
    for i, species in enumerate(df['species'].unique()):
        subset = df[df['species'] == species]
        ax.hist(subset[feature], bins=15, alpha=0.6, color=color_list[i], 
                label=species, edgecolor='black', linewidth=0.5)
    ax.set_xlabel(title, fontsize=13)
    ax.set_ylabel('Frequency', fontsize=13)
    ax.set_title(f'Distribution of {title}', fontsize=14, fontweight='bold')
    ax.legend(fontsize=10)

plt.tight_layout()
plt.show()
```

**Observations from Histograms:**
- **Petal Length** and **Petal Width** show clear bimodal distributions — Setosa is clearly separated from Versicolor and Virginica.
- **Sepal Width** is approximately normally distributed.
- **Sepal Length** shows moderate overlap between the species.

### 6.3 Box Plots — Detecting Outliers and Spread of Values

```python
# Box Plots for all features grouped by species
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Box Plots of Iris Features by Species', fontsize=18, fontweight='bold', y=1.01)

for idx, (feature, title) in enumerate(zip(features, titles)):
    ax = axes[idx // 2, idx % 2]
    sns.boxplot(x='species', y=feature, data=df, ax=ax, 
                palette=palette, width=0.5, linewidth=1.5,
                flierprops=dict(marker='o', markerfacecolor='red', markersize=6))
    ax.set_xlabel('Species', fontsize=13)
    ax.set_ylabel(title, fontsize=13)
    ax.set_title(f'Box Plot: {title} by Species', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

**Observations from Box Plots:**
- **Setosa** has the smallest petal length and petal width — it is very compact and distinct.
- **Virginica** generally has the largest measurements across most features.
- **Sepal Width** shows some outliers for Virginica.
- There is noticeable **overlap** in sepal measurements between Versicolor and Virginica, but petal measurements separate them better.

### 6.4 Correlation Heatmap

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
numeric_df = df.select_dtypes(include=[np.number])
correlation = numeric_df.corr()

sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0, 
            fmt='.2f', linewidths=2, linecolor='white',
            square=True, cbar_kws={'shrink': 0.8})
plt.title('Correlation Heatmap of Iris Features', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

**Observations from Correlation Heatmap:**
- **Petal Length** and **Petal Width** are highly correlated (0.96) — flowers with longer petals tend to have wider petals.
- **Sepal Length** correlates well with Petal Length (0.87) and Petal Width (0.82).
- **Sepal Width** has a weak negative correlation with other features, meaning it behaves differently.

### 6.5 Violin Plot — Distribution + Box Plot Combined

```python
# Violin Plots for a more detailed distribution view
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Violin Plots of Iris Features by Species', fontsize=18, fontweight='bold', y=1.01)

for idx, (feature, title) in enumerate(zip(features, titles)):
    ax = axes[idx // 2, idx % 2]
    sns.violinplot(x='species', y=feature, data=df, ax=ax, 
                   palette=palette, inner='box', linewidth=1.5)
    ax.set_xlabel('Species', fontsize=13)
    ax.set_ylabel(title, fontsize=13)
    ax.set_title(f'Violin Plot: {title} by Species', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

### 6.6 Summary Statistics by Species

```python
# Detailed statistics grouped by species
print("Summary Statistics by Species:")
print("=" * 60)
df.groupby('species').describe().T
```

```python
# Mean values for each feature grouped by species
print("Mean Values by Species:")
print("=" * 50)
species_means = df.groupby('species').mean()
print(species_means)

# Visualize the mean values
species_means.plot(kind='bar', figsize=(12, 6), colormap='Set2', edgecolor='black', linewidth=0.5)
plt.title('Mean Feature Values by Species', fontsize=16, fontweight='bold')
plt.xlabel('Species', fontsize=14)
plt.ylabel('Mean Value (cm)', fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.legend(title='Features', fontsize=11, title_fontsize=12)
plt.tight_layout()
plt.show()
```
